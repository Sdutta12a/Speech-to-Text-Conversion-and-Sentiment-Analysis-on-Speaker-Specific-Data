```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
data = pd.read_excel('speakers_text.xlsx')
```

```python
data.head()
```

```python
from textblob import TextBlob
```

```python
text = data['Text']
```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

# Function to categorize sentiment based on polarity
def categorize_sentiment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# Apply sentiment analysis to each text
data['polarity'], data['subjectivity'] = zip(*data['Text'].apply(get_sentiment))

# Categorize sentiment based on polarity
data['sentiment'] = data['polarity'].apply(categorize_sentiment)
pd.set_option('display.max_colwidth', None)  # To ensure text columns are fully visible
pd.set_option('display.expand_frame_repr', False)
# Display the results
print(data[['Speaker', 'Text', 'polarity', 'subjectivity', 'sentiment']])
```

```python
def map_emotion(polarity):
    if polarity > 0:
        return 'Happy/Pleasant'
    elif polarity < 0:
        return 'Sad/Angry'
    else:
        return 'Neutral/Nil'

# Categorize emotions based on polarity
data['emotion'] = data['polarity'].apply(map_emotion)

# Display the results
print(data[['Speaker', 'Text', 'polarity', 'subjectivity', 'sentiment', 'emotion']])
```

```python
from transformers import pipeline

# Load the pre-trained emotion detection model
emotion_classifier = pipeline('text-classification', model='j-hartmann/emotion-english-distilroberta-base')

# Apply the model to detect emotions
data['emotion'] = data['Text'].apply(lambda text: emotion_classifier(text)[0]['label'])

# Display the results
print(data[['Speaker', 'Text', 'polarity', 'subjectivity', 'sentiment', 'emotion']])
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Load the emotion detection model
emotion_classifier = pipeline('text-classification', model='j-hartmann/emotion-english-distilroberta-base')

# Step 2: Apply the model to detect emotions
data['Predicted_Emotion'] = data['Text'].apply(lambda text: emotion_classifier(text)[0]['label'])
data['Actual_Emotion'] = ['sadness','happiness','neutral','joy','joy','joy','anger','fear','neutral','joy','joy','joy','joy','disgust','sadness','joy','neutral','joy','joy','joy','joy','sadness','happiness','sadness','joy','fear','disgust','joy','sadness','surprise','sadness','joy','sad','neutral','fear','joy','joy','sadness','neutral']
# Step 4: Calculate the accuracy
accuracy = accuracy_score(data['Actual_Emotion'], data['Predicted_Emotion'])
print(f"Accuracy: {accuracy * 100:.2f}%")

# Step 5: Generate a detailed classification report (precision, recall, F1-score)
report = classification_report(data['Actual_Emotion'], data['Predicted_Emotion'])
print(report)

# Step 6: Display the DataFrame with actual and predicted emotions
print(data[['Speaker', 'Text', 'Actual_Emotion', 'Predicted_Emotion']])
```
